# Aula de Condicionais


## Condicionais

Condicionais são estruturas de programação que executam ações a partir de correponsdência de certas condições.

## If
If ("Se" em português) é uma estrura que executa uma ação caso certas condições determinadas sejam atendidas.

## Else
Else ("Se não" em português) é um estrutura que executa uma ação caso certas condições determinadas estabelecidas anteriormente (if) <b>não</b> sejam atendidas. 

## Elif
Elif (Else + If) é uma estrura que executa uma ação caso certas condições <b>distintas</b> daquelas determinadas anteriormente sejam atendias.


## And
And ("e" em português) é uma estrutura que permite determinar duas ou mais condições que devem ser simultâneamente atendidas para a execução de uma ação (Se x < 0 <b>e</b> y = 0:).

## Or
Or ("ou" em português) é uma estrutura que permite determinar duas ou mais condições de tal modo que no mínimo uma delas deve ser atendida para a execução de uma ação (Se x < 0 <b>ou</b> y = 0:)

## Exemplo:
Um fazendeiro possui uma fazenda de tomates e deseja vende-los de acordo com seu tamanho e maturação. Se os tomates forem grandes(g) e maduros ele os vendera por 2,00 reais a unidade. Se os tomates forem pequenos e maduros ou grandes e verdes ele deve os vender por 1,00 real a unidade. Caso o tomate for pequeno e verde ele deve os descartar.

```python
print("Qualidade do Tomate")
Maturação = input("Maturação(v/m): ")
Tamanho = input('Tamanho(p/g): ')

if Maturação == "m" and Tamanho == "g":
    print("Vender tomate a R$2,00.")
elif Maturação == "m" and Tamanho == "p" or Maturação == "v" and Tamanho == "g":
    print("Vender tomate a R$1,00")
else:
    print("Descartar tomate.")
```

# Exercício 1
Você aprendeu na escola a calcular o IMC (índice de massa corporal), porém, já cansou de ficar toda hora procurando uma calculadora no Google 🔍🙄
Mas agora você é DEV! 🧑💻 Faça a sua própria calculadora de IMC com um input para o usuário, a calculadora em si, a divisão por categorias e um output que dê o diagnóstico (dica: utilize as funções if, elif e else).

Use as informações abaixo como parâmetro:

Muito abaixo do peso <17

Abaixo do peso <18,5

Peso normal <25

Sobrepeso <30

Obesidade grau I <35

Obesidade grau II (severa) <40

Obesidade grau III (mórbida) >40

Lembrando que a fórmula do IMC é 𝐼𝑀𝐶 =
𝑃𝑒𝑠𝑜/𝐴𝑙𝑡𝑢𝑟𝑎²


```python
# Calculadora de IMC  (Índice de Massa Corporal)

# Entrada (input) de dados do usuário
height = float(input("Digite sua altura em metros:\n"))
weight = float(input("Digite seu peso em quilogramas:\n"))

# Calcula o IMC
imc = weight / (height ** 2)

# Separa as faixas de IMC e a variável result
if imc < 17:
	result = "Você está muito abaixo do peso"
elif imc < 18.5:
    result = "Você está abaixo do peso"
elif imc < 25:
    result = "Seu peso está normal"
elif imc < 30:
    result = "Você está com sobrepeso"
elif imc < 35:
    result = "Você tem obesidade grau I"
elif imc < 40:
    result = "Você tem obesidade grau II (severa)"
else:
    result = "Você tem obesidade grau III (mórbida)"

# Exibe o IMC calculado e o diagnóstico (variável result)
print(f"Seu IMC é {imc:.2f}")
print("Diagnóstico: " + result)
```

Eles sabem que o bebê não pode ter nome de uma pessoa que já existe na família, por isso fizeram uma lista com todos os nomes que já existem na família. Além disso, eles querem que o nome não seja grande, então precisa ter menos de nove caracteres. Seu desafio é ajudar o casal a escolher um nome para a criança.

##### Instruções:
        
        - O programa irá perguntar se a sugestão de nome é de menina ou menino.
        - Se for de menino e seguir as regras estabelecidas, o programa irá imprimir o nome escolhido com a mensagem: "se for menina se chamará: " + nome
        - Se for de menina e seguir as regras estabelecidas, o programa irá imprimir o nome escolhido com a mensagem: "se for menino se chamará: " + nome
        

```python
nome_familia = ['Miguel', 'Arthur', 'Heitor', 'Davi', 'Bernardo', 'Gabriel', 'Pedro', 'Anthony', 'Matheus', 'Lucas','Helena', 'Alice', 'Laura', 'Sophia', 'Manuela', 'Cecília', 'Isabella', 'Luísa', 'Heloísa', 'Júlia', 'Elisa', 'Valentina', 'Lara', 'Maria Clara', 'Lívia']
sexo = input('Digite o sexo da criança:  (M ou F)')
if sexo == 'M':
    nome = input('Digite a sugestão de nome da criança: ')
    if nome not in nome_familia and len(nome)<=8:
        print(f'se for menino se chamará: {nome}')
    else:  
        print('O nome é inválido, tente outro nome')
elif sexo == 'F':
    nome = input('Digite a sugestão de nome da criança: ')
    if nome not in nome_familia and len(nome)<=8:
        print(f'se for menina se chamará: {nome}')
    else:
        print('O nome é inválido, tente outro nome')
```
