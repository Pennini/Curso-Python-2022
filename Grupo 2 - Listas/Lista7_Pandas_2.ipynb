```python
##### SEU CODIGO COMECA AQUI #####
import pandas as pd
import numpy as np
import random

ativo = pd.read_csv("ativos.csv", index_col = 0, parse_dates = True)
carteira = pd.read_csv("carteiras.csv", index_col = 5, parse_dates = ["data"])
```

**Exercício 1: Filtragem de dados**

Filtre o Dataframe a partir de 5 ações, de modo que essas ações podem ser escolhidas pela sua preferência ou de forma aleatória. Após a filtragem, salve em uma nova variável.

```python
##### SEU CODIGO COMECA AQUI #####
acoes = random.choices(ativo.columns, k=5)
acoes
```

```python
df_acoes = ativo.filter(items = acoes, axis = 1)
df_acoes
```

**Exercício 2: Retorno**

Utilizando o Dataframe filtrado no exercício anterior, calcule o retorno diário dessas 5 ações. Para isso, utilize o método `pct_change`. Além disso, mude o nome das colunas para o nome da ação seguido de uma hashtag, por exemplo, 'BBAS3#'.

salve um uma variável chamada `retorno`.

```python
##### SEU CODIGO COMECA AQUI #####
retorno = df_acoes.pct_change()

for i in retorno.keys():
    retorno[i + "#"] = retorno[i]
    del retorno[i]

retorno = pd.DataFrame(retorno)
retorno
```

**Exercício 3: Ordenar**

Utilizando o método `describe`, analise a média de retorno das ações. Após isso, organize os dados de tal modo que as ações sejam apresentadas de forma decrescente, isto é, a ação que teve a maior média de retorno para a  ação que teve a menor média de retorno.

  
   *Lembre-se de definir o parâmetro 'ascending' e 'by'.*

```python
##### SEU CODIGO COMECA AQUI #####
retorno.describe().sort_values(by='mean', axis=1, ascending=False)
```

**Exercício 4: Manipulação** 

Para a ação que obteve a maior média de retonos no périodo apresentado, crie uma coluna chamada `acima_da_media_x`, sendo o 'x' o nome da ação. Sendo assim, essa coluna deve conter valores booleanos, `True` caso a ação esteja com o preço acima da sua média e `False` caso o preço da ação seja menor ou igual a média. Feito isso, printe apenas os dados que são iguais a `True`. 

*Dessa forma, poderemos saber em quais dias que a ação obteve um retorno maior que a sua média*

```python
##### SEU CODIGO COMECA AQUI #####
acima_da_media = []
nome_acao = retorno.columns[0]
for dia in retorno[nome_acao]:
    if dia > -0.000083:
         acima_da_media.append(True)
    else:
        acima_da_media.append(False)

retorno[f'acima_da_media_{nome_acao}'] = acima_da_media
retorno[retorno[f'acima_da_media_{nome_acao}'] == True]
```

**Exercício 5: Proporção**

Após descobrir em quais dias que a ação teve um retorno maior que a sua média, calcule qual foi a proporção em que isso aconteceu, ou seja, a proporção de dias que esteve acima da média em relação ao total de dias.

*printe essa proporção da seguinte forma: 'A proporção de dias acima da média foi de X%', sendo 'X' a proporção. Arredonde em duas casa decimais o valor encontrado*

```python
##### SEU CODIGO COMECA AQUI #####
dias_acima = retorno[retorno[f'acima_da_media_{nome_acao}'] == True].count()
dias_total = retorno.count()

proporcao = dias_acima[f'acima_da_media_{nome_acao}'] / dias_total[f'acima_da_media_{nome_acao}']
proporcao
```
